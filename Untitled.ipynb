```python
import pandas as pd 
data = pd.read_csv("data.csv")
df = data.copy()
df.head()
```

```python
df.drop(["otel"],axis =1,inplace=True)
df.info()
```

```python
df
```

```python
for i in['konum','secenek']:
    print("Column: ",i)
    print(df[i].unique())
```

```python
from sklearn.preprocessing import LabelEncoder  #sayısal veri verme

LE = LabelEncoder()
label_dict = {}
for i in['konum','secenek']:
    df[i] = LE.fit_transform(df[i])
    label_dict[i] = {label:encoded_label for label,encoded_label in zip(LE.classes_,LE.transform
     (LE.classes_))}
```

```python
df
```

```python
label_dict
```

```python
df.isnull().sum()   #bos deger var mı
```

```python
X = df.drop("fiyat",axis =1)
y = df["fiyat"]

```

```python
 from sklearn.linear_model import LinearRegression, Ridge, Lasso ,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import time
import warnings
warnings.filterwarnings('ignore')

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=37)

models = [
    LinearRegression(), Ridge(), Lasso() ,ElasticNet(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    XGBRegressor()
]

scores_df = pd.DataFrame(columns=['Model','Score','Time'])

for model in models:
    start = time.time()
    model.fit(X_train,y_train)
    end = time.time()
    
    
    scores_df = scores_df._append({'Model':type(model).__name__,'Score':model.score(X_test,y_test),
      'Time':end-start},ignore_index=True)
    print(f"{type(model).__name__} - {model.score(X_test,y_test)}")
```

```python
scores_df
```
